```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the CSV file
df = pd.read_csv("data/research_papers_timeline.csv")
sns.set(style="whitegrid", color_codes=True)
# Set a Seaborn theme for better visuals
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
    "font.size": 14,
    # "pdf.fonttype": 3,
    "text.latex.preamble": r"\usepackage[T1]{fontenc}",
    })
# sns.set_theme(style="whitegrid", palette="pastel")
# sns.set(font_scale=1.2)

# Ensure consistent categories and colors
category_order = ["Web Crawlers", "Tracker Detection (non-ML)", "Tracker Detection (ML)"]
colors = sns.color_palette("pastel", n_colors=3)

# Group data by year and category
pivoted_data = df.groupby(['year', 'category']).size().unstack(fill_value=0)[category_order]

# Plot 1: Stacked bar chart with Web Crawlers below x-axis
fig, ax = plt.subplots(figsize=(16, 6))

# Plot 'Web Crawlers' below the baseline (simulate with no negatives)
ax.bar(pivoted_data.index, pivoted_data['Web Crawlers'], label='Web Crawlers', color=colors[0])
ax.bar(pivoted_data.index, pivoted_data['Tracker Detection (non-ML)'], bottom=pivoted_data['Web Crawlers'],
       label='Tracker Detection (non-ML)', color=colors[1])
ax.bar(pivoted_data.index, pivoted_data['Tracker Detection (ML)'],
       bottom=pivoted_data['Web Crawlers'] + pivoted_data['Tracker Detection (non-ML)'],
       label='Tracker Detection (ML)', color=colors[2])

# Customization
ax.axhline(0, color='black', linewidth=1)
ax.set_xlabel('Year', fontsize=16)
ax.set_ylabel('Number of Papers', fontsize=16)
ax.legend(title='Category', loc='upper left')
plt.xticks(sorted(df['year'].unique()))
plt.title('Distribution of Papers And Influential Contributions Over Time', fontsize=20)
plt.grid(axis='x')

plt.text(
    2016, 1,  # Position of the textbox (x, y)
    'OpenWPM',  # Text content
    fontsize=12,  # Font size
    ha='center',  # Horizontal alignment
    va='center',  # Vertical alignment
    bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round,pad=0.5')  # Box style
)

plt.text(
    2018, 0.5,  # Position of the textbox (x, y)
    'AdGraph',  # Text content
    fontsize=12,  # Font size
    ha='center',  # Horizontal alignment
    va='center',  # Vertical alignment
    bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round,pad=0.5')  # Box style
)

plt.text(
    2018, 6,  # Position of the textbox (x, y)
    'GDPR',  # Text content
    fontsize=12,  # Font size
    ha='center',  # Horizontal alignment
    va='center',  # Vertical alignment
    bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round,pad=0.5')  # Box style
)

plt.text(
    1999, 0.5,  # Position of the textbox (x, y)
    'ADEATER',  # Text content
    fontsize=12,  # Font size
    ha='center',  # Horizontal alignment
    va='center',  # Vertical alignment
    bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round,pad=0.5')  # Box style
)

plt.text(
    2018, 5,  # Position of the textbox (x, y)
    'Tranco',  # Text content
    fontsize=12,  # Font size
    ha='center',  # Horizontal alignment
    va='center',  # Vertical alignment
    bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round,pad=0.5')  # Box style
)

plt.text(
    2005, 6,  # Position of the textbox (x, y)
    'EasyList',  # Text content
    fontsize=12,  # Font size
    ha='center',  # Horizontal alignment
    va='center',  # Vertical alignment
    bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round,pad=0.5')  # Box style
)

plt.text(
    2006, 5,  # Position of the textbox (x, y)
    'EasyPrivacy',  # Text content
    fontsize=12,  # Font size
    ha='center',  # Horizontal alignment
    va='center',  # Vertical alignment
    bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round,pad=0.5')  # Box style
)

plt.text(
    2012, 0.5,  # Position of the textbox (x, y)
    'FourthParty',  # Text content
    fontsize=12,  # Font size
    ha='center',  # Horizontal alignment
    va='center',  # Vertical alignment
    bbox=dict(facecolor='white', alpha=0.8, edgecolor='black', boxstyle='round,pad=0.5')  # Box style
)

plt.tight_layout()
plt.savefig('paper_dist.pdf', format='pdf', bbox_inches='tight', pad_inches=0.1)
plt.show()
```
